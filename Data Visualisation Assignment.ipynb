```python
# Task 2
sns.histplot(data=ds, x="price",kde=True)
plt.show()
```

```python
# Task 3 
# Count records by cut type
cut_counts = ds["cut"].value_counts()
# Print actual values
print(cut_counts)
# Plot bar chart
sns.barplot(x=cut_counts.index, y=cut_counts.values)
sns.despine()
```

```python
# Task 4
count = ds.groupby('cut')['price'].mean()
print(count)
# used line plot which will enable us to see which has the highest and which has the least mean average easily
sns.barplot(x='cut', y='price',data=ds, ci= None, marker='o') 
```

```python
sns.barplot(x = 'cut', y = 'price',  data=ds)
```

```python
# Task 5 
# Create a crosstab table of cut and color
table = pd.crosstab(ds["cut"], ds["color"])
table
```

```python
# Plot the table as a line chart with markers
a = sns.lineplot(data=table,marker='s')
plt.show()
```

```python
# Task 6 
ds.groupby(['cut','color'])['carat'].mean()
```

```python
sns.lineplot(x='cut',y='carat',hue='color',data=ds,ci=None, marker="s")
```

```python
# Task  7
# Carat and Price relation 
sns.scatterplot(x='carat',y='price',hue='cut',data=ds)
```

```python
# Carat and x Dimension relation 
sns.scatterplot(x='carat',y='x',hue='cut',data=ds)
```
